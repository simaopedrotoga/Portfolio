Procedimentos:
    1. Calcular o intervalo de confiana para o valor mdio populacional
    2. Calcular o intervalo de confiana para a varincia populacional
    3. Calcular o intervalo de confiana para o desvio padro populacional

```python
# Calcular o intervalo de confiana para o valor mdio populacional
## Nvel de confiana - 95%; Nvel de significncia - 5%; Dimenso da mostra - Grande; desvio populacional - Conhecido 
### intervalo = [mdia da amostra +- z * desvio padro populacional/raiz quadrada da dimenso da amostra]
amostras_a_complt = 100
amostras_quant = 0
amostras_dimens = 50
z = 1.96 # Ir  tabela e encontrar o nmero em que a probabilidade  0.0025 (nvel de significncia/2 = 0.05/2), 
         #visto que a distribuio  simtrica
desvio_pop = 14.5858964509
intervalo_amostrais_media_final = []
while amostras_quant < amostras_a_complt:
    intervalo_amostrais_media = []
    x1 = np.random.choice(frequencia, amostras_dimens).mean()
    int1 = x1 - z * desvio_pop/np.sqrt(amostras_dimens)
    intervalo_amostrais_media.append(int1)
    x2 = np.random.choice(frequencia, amostras_dimens).mean()
    int2 = x2 + z * desvio_pop/np.sqrt(amostras_dimens)
    intervalo_amostrais_media.append(int2)
    intervalo_amostrais_media_final.append(intervalo_amostrais_media)
    amostras_quant += 1

# Calcular o intervalo de confiana para a varincia populacional
## Nvel de confiana - 95%; Nvel de significncia - 5%; Dimenso da mostra - Grande
amostras_a_complt = 100
amostras_quant = 0
amostras_dimens = 50
xn2_a = 71.42 # Ir  tabela e encontrar o nmero que tenha os graus de liberdade igual a 49 (N-1 = 50-1) 
              #e com uma probabilidade a 0.025 (nvel de significncia/2 = 0.05/2)
xn2_b = 32.357 # Ir  tabela e encontrar o nmero que tenha os graus de liberdade igual a 49 (N-1 = 50-1) 
               #e com uma probabilidade a 0.975 (1 - (nvel de significncia/2) = 1 - (0.05/2))
intervalo_amostrais_var_final = []
while amostras_quant < amostras_a_complt:
    intervalo_amostrais_var = []
    var1 = np.random.choice(frequencia, amostras_dimens).var()
    int1 = ((amostras_dimens - 1) * var1) / (xn2_a)
    intervalo_amostrais_var.append(int1)
    var2 = np.random.choice(frequencia, amostras_dimens).var()
    int2 = ((amostras_dimens - 1) * var1) / (xn2_b)
    intervalo_amostrais_var.append(int2)
    intervalo_amostrais_var_final.append(intervalo_amostrais_var)
    amostras_quant += 1

# Calcular o intervalo de confiana para o desvio padro populacional
## Nvel de confiana - 95%; Nvel de significncia - 5%; Dimenso da mostra - Grande
amostras_a_complt = 100
amostras_quant = 0
amostras_dimens = 50
xn2_a = 71.42 # Ir  tabela e encontrar o nmero que tenha os graus de liberdade igual a 49 (N-1 = 50-1) 
              #e com uma probabilidade a 0.025 (nvel de significncia/2 = 0.05/2)
xn2_b = 32.357 # Ir  tabela e encontrar o nmero que tenha os graus de liberdade igual a 49 (N-1 = 50-1) 
               #e com uma probabilidade a 0.975 (1 - (nvel de significncia/2) = 1 - (0.05/2))
intervalo_amostrais_desv_final = []
while amostras_quant < amostras_a_complt:
    intervalo_amostrais_desv = []
    var1 = np.random.choice(frequencia, amostras_dimens).var()
    int1 = ((amostras_dimens - 1) * var1) / (xn2_a)
    intervalo_amostrais_desv.append(np.sqrt(int1))
    var2 = np.random.choice(frequencia, amostras_dimens).var()
    int2 = ((amostras_dimens - 1) * var1) / (xn2_b)
    intervalo_amostrais_desv.append(np.sqrt(int2))
    intervalo_amostrais_desv_final.append(intervalo_amostrais_desv)
    amostras_quant += 1

print("--------------------------------------Intervalos de confiana para o valor mdio populacional---\
-------------------------------")
print("\n")
for x in intervalo_amostrais_media_final:
    print(x)
print("\n")
print("Intervalos que no cumprem os requisitos")
print("\n")
for x in intervalo_amostrais_media_final:
    if (x[0] > 65.73343383999999) or (x[1] < 65.73343383999999):
        print(x)
print("\n")
print("--------------------------------------Intervalos de confiana para a varincia populacional------\
------------------------------")
print("\n")
for y in intervalo_amostrais_var_final:
    print(y)
print("\n")
print("Intervalos que no cumprem os requisitos")
print("\n")
for x in intervalo_amostrais_var_final:
    if (x[0] > 212.748375277) or (x[1] < 212.748375277):
        print(x)
print("\n")
print("------------------------------------Intervalos de confiana para o desvio padro populacional------\
----------------------------")
print("\n")
for w in intervalo_amostrais_desv_final:
    print(w)
print("\n")
print("Intervalos que no cumprem os requisitos")
print("\n")
for x in intervalo_amostrais_desv_final:
    if (x[0] > 14.5858964509) or (x[1] < 14.5858964509):
        print(x)
print("\n")
print("Nota: Se formos a tirar amostras aleatrias repetidamente das 100 geradas, o valor real de cada \
parmetro estar dentro das 100 amostras 95% das vezes. Se sair outro valor,  normal pois existe 5% de erro")   
```

```python
# Encontrando o nmero ideal para o hiperparmetro de clusters com o mtodo Elbow
distortions = []
inertias = []
clusters = range(1, 10)
for cluster in clusters:
    modelo = sklearn.cluster.KMeans(n_clusters = cluster).fit(df)
    
    # Utilizando Distortion
    distortions.append(sum(np.min(cdist(df, modelo.cluster_centers_,
                                        'euclidean'), axis=1)) / df.shape[0])
    # Utilizando Inertia
    inertias.append(modelo.inertia_)

fig = plt.figure(figsize = (8, 4), dpi = 100)

ax1 = fig.add_subplot(1,2,1)
ax1.plot(clusters, distortions, color = "brown", zorder = 100)
ax1.set_title('O Mtodo Elbow usando Distortion')
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('Distortion')
ax1.grid(color = '0.90')


ax2 = fig.add_subplot(1,2,2)
ax2.plot(clusters, inertias, color = "brown", zorder = 100)
ax2.set_title('O Mtodo Elbow usando Inertia')
ax2.set_xlabel('Number of Clusters')
ax2.set_ylabel('Inertia')
ax2.grid(color = '0.90')

plt.show()
```

### <font color=green>Abordagem Classification</font>

Na procura do melhor resultado possvel, ir ser utilizado um algoritmo ensemble, mais especificamente, o algoritmo Random Forest.

```python
# Definir o conjunto de dados
X = df[['2.4 - Sabe o que significa "Fake News"?', 
         "2.4.1 - Se sim, qual das seguintes opes corresponde ao seu significado?",
         '2.4.2 - O tema das "Fake News"  um assunto de extrema importncia e que prejudica a divulgao do conhecimento.',
         "2.5 - Na maioria dos casos, como se comporta quando se depara com uma notcia?",
        "2.6 - Com que frequncia utiliza os seguintes critrios como forma de analisar uma notcia? [Verifico a fonte da notcia]",
        "2.6 - Com que frequncia utiliza os seguintes critrios como forma de analisar uma notcia? [Verifico outras fontes para confirmar a notcia]",
        "2.6 - Com que frequncia utiliza os seguintes critrios como forma de analisar uma notcia? [Verifico o autor da notcia]",
        "2.6 - Com que frequncia utiliza os seguintes critrios como forma de analisar uma notcia? [Verifico o aspeto da pgina (caso se trate de uma notcia online)]",
        "2.6 - Com que frequncia utiliza os seguintes critrios como forma de analisar uma notcia? [Verifico a data da notcia]",
        "2.6 - Com que frequncia utiliza os seguintes critrios como forma de analisar uma notcia? [Consulto uma pessoa que esteja mais  vontade com o assunto da notcia]",
        "2.7 - Qual a fonte de informao que mais utiliza na satisfao de uma necessidade de informao geral de fcil acesso?",
        "2.8 - Qual a fonte de informao que mais utiliza na satisfao de uma necessidade de informao especfica de difcil acesso?",
        "2.9 - Costuma utilizar operadores boleanos (AND, NOT e OR) para restringir as suas pesquisas?",
        "2.10 - Ordene as seguintes opes em termos de relevncia na busca de uma certa informao  [Fiabilidade]",
        "2.10 - Ordene as seguintes opes em termos de relevncia na busca de uma certa informao  [Relevncia]",
        "2.10 - Ordene as seguintes opes em termos de relevncia na busca de uma certa informao  [Atualizao]",
        "2.10 - Ordene as seguintes opes em termos de relevncia na busca de uma certa informao  [Completa]",
        "2.10 - Ordene as seguintes opes em termos de relevncia na busca de uma certa informao  [Explicita]",
        "4 - Ter mais ateno agora quando for ler uma notcia?"]].to_numpy()
y = df['Nota'].to_numpy()

# Valores a avaliar
desempenhos = []
k_values = [1, 2, 3, 4]
for k in k_values:
    # Criando um modelo de classificao
    modelo = sklearn.ensemble.RandomForestClassifier()
    
    # Criando um conjunto de dados de treino mais balanceado
    over = SMOTE(sampling_strategy = {2.0: 55, 3.0: 81, 4.0: 82, 5.0: 20}, k_neighbors = k)
    under = RandomUnderSampler(sampling_strategy = {2.0: 55, 3.0: 60, 4.0: 61, 5.0: 20})
    
    # Preparando o processo a ser executado
    passos = [('over', over), ('under', under), ('modelo', modelo)]
    pipeline = Pipeline(steps = passos)
    
    # Treinando o modelo e fazendo previses
    cv = sklearn.model_selection.RepeatedStratifiedKFold(n_splits = 6, n_repeats = 3, random_state = 1)
    scores = sklearn.model_selection.cross_val_score(pipeline, X, y, scoring = 'accuracy', cv = cv, n_jobs = -1)
    print(scores)
    desempenhos.append(np.mean(scores))
```

### <font color=green>Validation</font>

Para verificar a qualidade do modelo, ser utilizado o acerto

```python
vizinho = 0
for desempenho in desempenhos:
    print('> k=%d, Mean Accuracy: %.3f' % (vizinho + 1, desempenho))
    vizinho += 1
```
